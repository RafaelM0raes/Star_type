Temperature = Temperatura (Kelvin)

L = Luminosidade Relativa (L / Lo)

R = Raio Relativo (R / Ro)

Ro = 6.9551 x 10^8 m (Raio Médio do Sol)

A_M = Magnitude Absoluta (Mv)

Color = Cor dentro do espectro geral de cores

Spectral_Class =  O,B,A,F,G,K,M / SMASS 

Type = Tipos de Estrelas (0 - Red Dwarf, 1 - Brown Dwarf, 2 - White Dwarf, 3 - Main Sequence , 4 - Super Giants, 5 - Hyper Giants) -> Target da set

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X = star_modificado.drop(columns=star_modificado.columns[4:7])
X
```

```python
y = star_modificado.drop(columns=star_modificado.columns[0:6])
y
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size= 0.33, random_state= 42)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
vizinhos = KNN(n_neighbors=5)
vizinhos.fit(X_train, y_train)
```

```python
y_pred = vizinhos.predict(X_test)
```

```python
print(y_pred)
```

```python
print(y_test)
```

```python
vizinhos.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
